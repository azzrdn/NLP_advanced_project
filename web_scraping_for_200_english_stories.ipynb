```python
# kütüphaneleri import etme
from bs4 import BeautifulSoup
import requests
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
page = requests.get("https://englishlearning.webgarden.es/menu/1st-and-2nd-eso-year/easy-reading-texts")
soup = BeautifulSoup(page.content, "html.parser")
```

```python
links = soup.find_all("span", style="font-size:10.0pt;line-height:115%")
```

```python
title = []
link = []
for i in range(len(links)-1):
    aziz1 = soup.find_all("span", style="font-size:10.0pt;line-height:115%")[i].text
    aziz2 = soup.find_all("span", style="font-size:10.0pt;line-height:115%")[i].a.get("href")
    print(aziz1)
    print(aziz2)
    title.append(aziz1)
    link.append(aziz2)
```

```python
for i in range(160,201):
    sites = "http://www.eslfast.com/supereasy/se/supereasy"+str(i)+".htm"
    link.append(sites)
    print(i)
    print(sites)
```

```python
titles = []
contents = []
for site in link:
    page = requests.get(site)
    soup = BeautifulSoup(page.content, "html.parser")
    title = soup.find_all("h1")[0].text.replace("\n ", "")
    print(title)
    titles.append(title)
    content = soup.find("div", class_="contain").text.replace("\n", " ")
    contents.append(content)
```

```python
df = pd.DataFrame({"title":titles, "text":contents})
```
