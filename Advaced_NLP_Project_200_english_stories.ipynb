## Stopword larÄ± temizleme

```python
import stopwords_aziz as aziz
```

```python
stopwords_en = set(aziz.stopwords_en)
```

```python
from nltk.tokenize import word_tokenize
```

```python
filtered_text = [word for word in word_tokenize(text) if word.lower() not in stopwords_en]
filtered_text
```

```python
#7. Bir metinde kelime anlamÄ±na uygun doÄŸru etiketleri bulma:
import nltk
# Currently, NLTK pos_tag only supports English and Russian
#word_tokens = nltk.word_tokenize(text)
#pos_tags = nltk.pos_tag(word_tokens, lang='eng') 
pos_tags = nltk.pos_tag(filtered_text)

adverb = []
noun = []
adjec = []
verb = []


for word, pos_tag in pos_tags:
    wordnet_tag = nltk.pos_tag([word])[0][1]
    if wordnet_tag.startswith("N"):
        noun.append(word)
        print(word, ":", "noun")
    elif wordnet_tag.startswith("V"):
        verb.append(word)
        print(word, ":", "verb")
    elif wordnet_tag.startswith("J"):
        adjec.append(word)
        print(word, ":", "adjective")
    elif wordnet_tag.startswith("R"):
        print(word, ":", "adverb")
        adverb.append(word)
    else:
        print(word, ":", "other")
```
